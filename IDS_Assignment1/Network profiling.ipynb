```python
import matplotlib.pyplot as plt
import seaborn as sns
def EDA(data_frame, graphs_details):
    figure, axes = plt.subplots(1, 3, figsize=(10,3))
    figure.tight_layout()
    counter = 0
    for index, feature in enumerate(graphs_details):
        if index % 3 == 0 and index != 0:
            figure, axes = plt.subplots(1, 3, figsize=(10,3))
            figure.tight_layout()
            counter = 0
        if feature['column_type'] == 'cat':
            df = data_frame.groupby(feature['column_name']).count()
            df.iloc[:,0].plot(kind=feature['graph_type'], ax=axes[counter])
        else:
            data_frame[feature['column_name']].plot(kind=feature['graph_type'], ax=axes[counter])
        counter += 1
        
```

```python
def outlier_suppression_skewed(data_frame, features):
    df = data_frame.describe()
    df = df[(df.index == '25%') | (df.index == '75%')]
    for feature in features:
        col = feature['column_name']
        q1 = df[col][0]
        q3 = df[col][1]
        IQR = q3 - q1
        data_frame.loc[(data_frame[col] > q3 + 1.5*IQR), col] = q3 + 1.5*IQR
        data_frame.loc[(data_frame[col] < q1 - 1.5*IQR), col] = q3 - 1.5*IQR
    return data_frame
        
```

```python
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]], drop_first= True)
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return(res) 

features_to_encode = ['flag', 'is_guest_login', 'is_host_login', 'protocol_type', 'root_shell', 
 'su_attempted', 'wrong_fragment', 'service', 'hot', 'logged_in']

for feature in features_to_encode:
    clean_df = encode_and_bind(data, feature)
```

```python
from sklearn.preprocessing import StandardScaler
# Separating out the features
x = clean_df.drop(['class'], axis=1).values
# Separating out the target
y = clean_df['class'].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2']
             )
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
finalDf = pd.concat([principalDf, clean_df['class']], axis = 1)
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [1, 0]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['class'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             )
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans
def k_means(df):
    kmeans_kwargs = {
         "init": "k-means++",
         "n_init": 10,
         "max_iter": 100,
         "random_state": 42,
     }
    # A list holds the silhouette coefficients for each k
    silhouette_coefficients = []
    ari_means_lst = []
    sse = []

    # Notice you start at 2 clusters for silhouette coefficient
    for k in range(2, 20):
        kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
        kmeans.fit(df)
        sse.append(kmeans.inertia_)
        ari_kmeans = adjusted_rand_score(y.reshape(22544), kmeans.labels_)
        ari_means_lst.append(ari_kmeans)
        score = silhouette_score(x, kmeans.labels_)
        silhouette_coefficients.append(score)
    return (kmeans, sse, silhouette_coefficients, ari_means_lst)
```

```python
original_featured_cluster = k_means(x)
pca_reduced_features_cluster = k_means(principalDf)
```

```python
figure, axes = plt.subplots(1, 1, figsize=(15,5))
axes.plot(range(2, 20), pca_reduced_features_cluster[2], label = "silhouette_coefficients")
axes.plot(range(2, 20), pca_reduced_features_cluster[3], label = "ari_means_lst")
axes.set_title("PCA feature reduced clustering")
axes.legend()
```
